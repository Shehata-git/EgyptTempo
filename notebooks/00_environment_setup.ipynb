python setup, upload .tar beforehand

```python
%%bash
# 1. Install uv directly from Astral (bypassing apt entirely)
curl -LsSf https://astral.sh/uv/install.sh | sh

# Add uv to the current bash path
export PATH="/root/.cargo/bin:$PATH"

# 2. Extract the uploaded project files
# (Change 'project.tar.gz' if your uploaded file has a different name)
tar -xzf project.tar.gz

# 3. Sync dependencies using uv
echo "Syncing dependencies..."
uv sync

echo "Environment ready."
```

system setup for mongo and hdf

```python
%%bash
# 1. Download and extract the official MongoDB binary (Colab compatible)
echo "Pulling MongoDB binaries..."
wget -qO- https://fastdl.mongodb.org/linux/mongodb-linux-x86_64-ubuntu2204-7.0.5.tgz | tar xz

# 2. Create the strict data directory MongoDB requires
mkdir -p /data/db

# 3. Start the daemon in the background (nohup prevents it from hanging the cell)
echo "Starting mongod daemon..."
nohup ./mongodb-linux-x86_64-ubuntu2204-7.0.5/bin/mongod --dbpath /data/db > mongod.log 2>&1 &
```
